```python
import eiapy as EIA
from meteostat import Hourly, Point, Stations
import pandas as pd
import numpy as np
from datetime import datetime
import pytz
import matplotlib.pyplot as plt
```

### Get the CO Power Demand data

```python
# Some info on the PowerDemand data
cat = EIA.Category(3389994)
cat.get_info()
```

```python
# All the data with metadata
series_id = cat.get_info()['category']['childseries'][1]['series_id']
all_data = EIA.Series(series_id).get_data(all_data=True)
all_data['series']
```

```python
# Load data into a pandas dataframe
power = pd.DataFrame(all_data['series'][0]['data'], columns=['Time','PowerDemand'])
power['Time'] = pd.to_datetime(power['Time'], infer_datetime_format=True)
power.head()

# Set the last time to use (should make it as current as possible before submitting)
end_time = datetime(2022,4,30,18,0,0)
tz = pytz.timezone('US/Mountain')       # add timezone info
aware = tz.localize(end_time, is_dst=None)
power = power[power['Time'] < aware]

# Save times as timezone aware in datetime format
times = [t.replace(tzinfo=None) for t in power.Time]     # first, save the time as unaware datetime
power['time_unaware'] = times
power['time_unaware'] = pd.to_datetime(power['time_unaware'])

times = [tz.localize(t) for t in power.time_unaware]     # add timezone info to unaware time
power['time'] = times

# Clean up the dataframe and rearrange so earliest time first
power = power.drop(['Time','time_unaware'], axis=1)
power = power[::-1].reset_index(drop=True)
power_length = power.shape[0]
print(power.shape)
power.head()
```

```python
# Test plot
power.plot(x='time', y='PowerDemand')
```

```python
# Save start and end times
start_date = power.iloc[0,1].strftime('%Y-%m-%d')
end_date = power.iloc[-1,1].strftime('%Y-%m-%d')
start_time = power.iloc[0,1]
end_time = power.iloc[-1,1]

print('Start: %s' %(start_time) )
print('End:   %s' %(end_time) )
```

### Get temperature data for cities and merge with power data

```python
# Load CO city data
cities = pd.read_csv('./city_data/top_20_CO_cities.csv')
cities = cities.drop(['Unnamed: 0'], axis=1)
cities['percent_pop'] = cities['population'] / cities['population'].sum()
cities.head()
```

```python
stations = Stations()

for i in range(len(cities)):
    # Get coordinates for each city
    city_name = cities.iloc[i,0]
    lat = cities.iloc[i,1]
    long = cities.iloc[i,2]
    
    # Use elevation of the nearest station to that city
    stations = stations.nearby(lat, long)
    station = stations.fetch(1)
    elevation = station.iloc[0,7]
    
    # Use Point in meteostat to get hourly data
    city = Point(lat, long, elevation)
    data = Hourly(city, start_time.replace(tzinfo=None), end_time.replace(tzinfo=None), timezone='US/Mountain')
    data = data.normalize()
    weather = data.fetch()
    
    # Only save temperatures and population
    temp = weather.drop(['dwpt', 'rhum', 'prcp', 'snow', 'wdir', 'wspd', 'wpgt', 'pres', 'tsun', 'coco'], axis=1)
    temp = temp.reset_index(level='time')
    temp['percent_pop'] = cities.iloc[i,4]
    
    # Merge temperature data with power data
    print('Merging temperature data for %s...' %(city_name))
    cols = {}
    for col in temp.columns:
        if col != 'time':
            cols[col] = col + '_' + city_name
            
    power = power.merge(temp, how='outer', on='time')
    power = power.rename(columns=cols)
    power = power[power.index < power_length]

print(power.shape)
power.head()
```

### Save data as CSV

```python
# Interpolate to fill in missing data
for col in power.columns:
    if power[col].isna().sum() != 0:
        print('Filling missing values in %s...' %(col))
        power[col] = power[col].interpolate()
```

### Get forecast data

```python
forecast_start = datetime(2022,4,30,18,0,0)
forecast_end = datetime(2022,5,1,18,0,0)
forecast_dates = pd.date_range(forecast_start, forecast_end, freq='H')
forecast = pd.DataFrame(data=forecast_dates, columns=['time'])

times = [tz.localize(t) for t in forecast.time]     # add timezone info to unaware time
forecast['time'] = times
print(forecast.dtypes)
forecast.head()
```

```python
stations = Stations()

for i in range(len(cities)):
    # Get coordinates for each city
    city_name = cities.iloc[i,0]
    lat = cities.iloc[i,1]
    long = cities.iloc[i,2]
    
    # Use elevation of the nearest station to that city
    stations = stations.nearby(lat, long)
    station = stations.fetch(1)
    elevation = station.iloc[0,7]
    
    # Use Point in meteostat to get hourly data
    city = Point(lat, long, elevation)
    data = Hourly(city, forecast_start, forecast_end, timezone='US/Mountain')
    data = data.normalize()
    weather = data.fetch()
    
    # Only save temperatures and population
    temp = weather.drop(['dwpt', 'rhum', 'prcp', 'snow', 'wdir', 'wspd', 'wpgt', 'pres', 'tsun', 'coco'], axis=1)
    temp = temp.reset_index(level='time')
    temp['percent_pop'] = cities.iloc[i,4]
    
    # Getting temperature data
    print('Getting forecast temperature data for %s...' %(city_name))
    cols = {}
    for col in temp.columns:
        if col != 'time':
            cols[col] = col + '_' + city_name
            
    forecast = forecast.merge(temp, how='outer', on='time')
    forecast = forecast.rename(columns=cols)
    
    
print(forecast.shape)
forecast.head()    
```

```python
# Interpolate to fill in missing data
for col in forecast.columns:
    if forecast[col].isna().sum() != 0:
        print('Filling missing values in %s...' %(col))
        forecast[col] = forecast[col].interpolate()
```

```python
# Create single time series of temperature weighted by population
weighted_temp = np.zeros(forecast.shape[0])
for col in forecast.columns:
    if col not in ['PowerDemand', 'time'] and not col.startswith('percent_pop'):
        city = col.split('_')[1]
        weighted_temp += forecast[col] * forecast['percent_pop_' + city]
        
forecast['WeightedTemp'] = weighted_temp
for col in forecast.columns:
    if col not in ['time', 'EnergyDemand', 'WeightedTemp']:
        forecast = forecast.drop([col], axis=1)
        
forecast.head()
```

```python
forecast.to_csv('./forecast.csv', index=False)
```
