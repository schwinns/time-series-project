```python
import pandas as pd
```

```python
cities = pd.read_csv('./uscities.csv')
print(cities.shape)
cities.head()
```

```python
CO = cities[cities.state_id == 'CO']
print(CO.shape)
CO.head()
```

```python
top_20 = CO.head(20)
top_20 = top_20.drop(['city_ascii', 'state_id', 'state_name',
                      'county_fips', 'county_name', 'density',
                      'source', 'military', 'incorporated',
                      'timezone', 'ranking', 'zips', 'id'], axis=1)
top_20 = top_20.reset_index(drop=True)
top_20.head()
```

```python
top_20.to_csv('top_20_CO_cities.csv')
```
